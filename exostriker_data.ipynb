```python
import exostriker.lib.RV_mod as rv
import pandas as pd

found_periods = []
found_amps = []
p_error = []
k_error = []
star_ID = []
for i in range(1,953):
    fit = rv.signal_fit(name="RV_Star{}".format(i)) #creates the "fit" object that
    fit.add_dataset("RV_Star{}.txt".format(i), "/Users/elizabethgonzalez/Desktop/ballard/TESS_Files/RV_Star{}.txt".format(i),0.0,0.0) # add the data file

    #fit.use.update_use_planet_params_one_planet(0,True,True,False,True,True,False,False)
    fit.fitting(outputfiles=[1,1,1], doGP=False,  minimize_fortran=True,
    minimize_loglik=False, amoeba_starts=20, print_stat=False)

    #Run GLS once 
    rv.run_gls_o_c(fit,fend=1.5, as_main = True)
    rv.run_gls_o_c(fit,fend=1.5)

    #now lets find the planets in our data!
    fit.auto_fit_max_pl = 8
    fit.auto_fit_FAP_level = 0.001 # this corresponds to FAP = 0.1%. GLS power with FAP below that level we take as planet candidate.
    fit.auto_fit_allow_ecc = True # otherwise will look only for circular planets
    fit = rv.find_planets(fit,fend=1.5)

    #Lets print the best fit params
    periods = fit.P
    periods_list = list(periods.values())
    amps = fit.K
    amps_list = list(amps.values())
    periods_array = [i for i in periods_list if i != 0]
    amps_array = [i for i in amps_list if i != 0]
    found_periods.extend(periods_array)
    found_amps.extend(amps_array)

    P_err_list = list(fit.P_err.values())
    K_err = fit.K_err
    K_err_list = list(fit.K_err.values())
    for j in range(0, len(P_err_list)):
        if P_err_list[j][0] != 0:
            p_error.append(P_err_list[j][0])
            k_error.append(K_err_list[j][0])
    
    number_planets = len(periods_array)
    star_ID.extend([int(i)]*number_planets)
```

```python
title = ['Star', 'Found_Period', 'Period_Error', 'Found_Amp', 'Amp_Error']     

length = len(found_periods)

df = pd.DataFrame(index=range(length), columns=title)
df['Star'] = star_ID[0:length]
df['Found_Period'] = found_periods[0:length]
df['Period_Error'] = p_error[0:length]
df['Found_Amp'] = found_amps[0:length]
df['Amp_Error'] = k_error[0:length]

df.to_pickle("found_data.pkl")
```

```python
import pandas as pd
import numpy as np
df = pd.read_pickle("found_data.pkl")
df_true = pd.read_pickle("true_data.pkl")
print(df, "\n", df_true)

df_true["Exo_Detection"] = np.nan
df_true["Found_Period"] = np.nan
df_true["Period_Error"] = np.nan
df_true["Found_Amp"] = np.nan
df_true["Amp_Error"] = np.nan


for index, row in df_true.iterrows():
    star = df_true.iloc[index]['Star']
    period = df_true.iloc[index]['True_Period']
    exostriker = df.loc[df['Star'] == star]
    for i in range(0, len(exostriker)):
        found_per = exostriker.iloc[i]['Found_Period']
        if found_per < period + 0.1 and found_per > period - 0.1:
            df_true.at[index, "Exo_Detection"] = 1 #found
            df_true.at[index, "Found_Period"] = exostriker.iloc[i]['Found_Period']
            df_true.at[index, "Period_Error"] = exostriker.iloc[i]['Period_Error']
            df_true.at[index, "Found_Amp"] = exostriker.iloc[i]['Found_Amp']
            df_true.at[index, "Amp_Error"] = exostriker.iloc[i]['Amp_Error']

pd.set_option("display.max_rows", None, "display.max_columns", None)
print(df_true)
```
